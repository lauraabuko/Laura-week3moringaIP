```python
td_clean = td_combine.drop(columns = ['country_a' ,	'country_b' ,'cell_on_site',"site_id",])
td_clean.head()
```

```python
# DROPPING DUPLICATES
td_cdrcombine = td_cdrcombine.drop_duplicates()
td_cdrcombine.head
```

```python
#COMBINING WHOLE TABLE
td_overall = pd.concat([td_cdrcombine,td_clean],axis=0)
td_overall.head

```
